## Official Course Description:
The course provides an early exposure to Earth data and their handling. The wide range of remotely or in-situ collected datasets will be introduced and examples provided, as well as their relevance for Earth science disciplines. Three main components of the course are included: a) an overview of the range of data types, standards, and formats commonly use in Earth science and exemplary data sources and data search, discovery, retrieval and inspecting options, b) exemplary Earth science data collection activities, and c) data handling and analysis basics, covering either archive or field-collected data. Geospatial and temporal data of interest and relevance for Earth and Marine sciences will be described, collected, inspected and analyzed with available open source tools.

### Intended Learning Outcomes (ILOs) 

• Exposure to the range of data typicaly use in Earth and Marine Science 
• Familiarisation with the open source tools available to analyse Earth Science data

# Unix and terminal (a quick recap ..)

You can find a terminal from the GUI of any modern operating unix-like operating system. It is used to launch commands, either with or without option, e.g.

with option:

```
user@host~> ls -lah
```

or without: 

```
user@host~> ls
```

The prompt can vary, but it typically shows your user ID, host and/or path:

```
user@host~> 
```


## Navigating directories - cd
When entering a terminal, typically you are in your ```home``` directory, which is also indicated with ```~``

change of directories can be achieved with the ```cd``` command, e.g.

```
user@host~> cd ~/de-earth/
user@host~/de-earth> 
```

## Listing files - ls
Files eventually present in a directory can be listed with ```ls```.
```
user@host~/de-earth> ls
image1.jpg test1.txt
```

## Figure out current directory - pwd
In order to figure out where you are in the filesystem, you can use ```pwd``` (print working directory..):

```
user@host~/de-earth> pwd
/Users/myname/de-earth
```

## Visualizing directory content and tree - tree

```
user@host~> cd ~/de-earth/
user@host~/de-earth> tree
.
├── image1.jpg
├── test2renamed.txt
├── test3-copy.txt
├── test3.txt
└── testdir
    └── test1.txt

1 directory, 5 files
```

## Wildcards - *
If you know only part of a filename, using ```*``` can be very handy. ```*``` means everything, and for example ```*.txt``` means everything with an extension ```.txt```, e.g.:

```
user@host~/de-earth> ls *txt
test1.txt test2.txt test3.txt
```

one can also use option, e.g. to list details of each files, ```ls``` accepts, among others, the ```-l``` option, i.e.:

```
user@host~/de-earth> ls -l *txt
-rw-r--r--  1 user  staff  15 Jan 14 09:15 test1.txt
-rw-r--r--  1 user  staff  20 Jan 14 09:18 test2.txt
-rw-r--r--  1 user  staff  48 Jan 14 09:18 test3.txt
```


# Creating directories and files

A directory/folder can be created with ```mkdir```, e.g. and removed (only if empty, otherwise one needs another command) with ```rmdir```.

```
user@host~/de-earth> mkdir testdir

user@host~/de-earth> ls
image1.jpg test1.txt  test2.txt  test3.txt  testdir
```

## Copy - cp
Files are copied with ```cp``` e.g.

```
user@host~/de-earth> cp test3.txt test3-copy.txt

user@host~/de-earth> 
```

## Move - mv
Moving somewhere or renaming a file uses the same command: ```mv```, as an example:

```
user@host~/de-earth> ls
image1.jpg     test2.txt      test3.txt
test1.txt      test3-copy.txt testdir

user@host~/introearthdata2018> mv test1.txt testdir/

user@host~/introearthdata2018> ls
image1.jpg     test2.txt      test3-copy.txt test3.txt      testdir

user@host~/introearthdata2018> ls testdir/
test1.txt
```

or, for renaming

```
user@host~/de-earth> ls
image1.jpg     test2.txt      test3-copy.txt test3.txt      testdir

user@host~/de-earth> mv test2.txt test2renamed.txt

user@host~/de-earth> ls
image1.jpg       test3-copy.txt   testdir
test2renamed.txt test3.txt
```

## Using the shell as a calculator

Python will be much better, but this is also possible. 

E.g. one can multiply 23 x 453 with bash calculator and redirect the result into another text file "calc.txt".

Tip: bash calculator is ``bc``. Piping is done with ```|```. Redirecting is done with ```>```.

```
user@host~> echo '23 * 453' | bc
user@host~> 10419
```

In order to redirect to a file:

```
user@host~> echo '23 * 453' | bc > result.txt
user@host~> more result.txt 
10419
```

### module vs. package
from http://stackoverflow.com/questions/7948494/whats-the-difference-between-a-python-module-and-a-python-package

A module is a single file (or files) that are imported under one import and used. e.g.

```>>> import my_module```

A package is a collection of modules in directories that give a package hierarchy.

```>>> from my_package.timing.danger.internets import function_of_love```

# Simple x,y plot


````python
# A simple plot

# I comment below "#plt.xkcd() ", as it makes all very slow (due to the deformation of each and every line!). 
# Figure looks nice with it, but that slows complex polygon rendering quite a lot!
# use at own risk!

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([10,20,50,76,245])

markers_on = [1,2,3,4,5]

# you can uncomment at your own risk..
#plt.xkcd()  

# figure
fig = plt.figure()

# subplot
ax1 = fig.add_subplot(111)

# plot line
ax1.plot(x, y, 'r', color='b')

# plot markers on top
ax1.plot(x, y, marker='o', color='r', ls='')

plt.xlabel('X label')
plt.ylabel('Y label')

# set x range
plt.xlim([0.5,5.5])

# tik location
plt.xticks([1, 2, 3, 4, 5])

# label rotation (try that out..)
locs, labels = plt.xticks()
plt.setp(labels, rotation=15)

# labels positioning and customising
fig.canvas.draw()

# show plot inline
plt.show()
````
